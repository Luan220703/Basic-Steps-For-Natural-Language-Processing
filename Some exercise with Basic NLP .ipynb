```python
import re

# Regular expression for arithmetic expression
arithmetic_expr_pattern = r'^\d+(\*\d+|\+\d+)*$'

# Regular expression for determiner
determiner_pattern = r'^(a|an|the)$'

# Test strings
test_arith = ["5*6+9","abc"]
test_deter = ["a","an","the"]

# Test arithmetic expressions
print("Arithmetic Expressions:")
for expr in test_arith:
    if re.match(arithmetic_expr_pattern, expr):
        print(f"{expr} matches the pattern")
    else:
        print(f"{expr} does not match the pattern")

# Test determiners
print("\nDeterminers:")
for word in test_deter:
    if re.match(determiner_pattern, word):
        print(f"{word} is a determiner")
    else:
        print(f"{word} is not a determiner")
```

```python
# Q3
from nltk import FreqDist
import re
def shorten(text, n):
  word_list = re.findall(r'\b\w+\b', text)
  freq = [i[0] for i in FreqDist(word_list).most_common(n)]
  res = []
  for word in word_list:
    if word not in freq:
      res.append(word)
  return ' '.join(res)
Text = 'Write a function shorten(text, n) to process a text, omitting the n most frequently occuring words of the text. How readable is it?'
shorten(Text, 4)
```

```python
import re
import nltk
from nltk.corpus import stopwords, gutenberg

def frequently_words(text: str, stop_words: set, number_words: int=50) -> dict:

    dict_words = {}

    word_list_from_text = re.split(r'[ \n]+', text)

    for word in word_list_from_text:

        if word not in stop_words:

            if word not in dict_words:
                dict_words[word] = 1
            else:
                dict_words[word] += 1

    dict_words = dict(sorted(dict_words.items(), key=lambda item: item[1], reverse=True))
    top_words = {k: v for k, v in list(dict_words.items())[:number_words]}

    return top_words

if __name__ == "__main__":

    #Create stopwords;
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    print(type(stop_words))

    #Create dataset from nltk;
    nltk.download('gutenberg')
    book_names = gutenberg.fileids()
    selected_book = 'austen-emma.txt'
    book_text = gutenberg.raw(selected_book)

    #Show answer;
    answers = frequently_words(book_text, stop_words)
    print(answers)
```

```python
def words_not_in_vocab(text, vocabulary):
    # Convert text and vocabulary lists to sets for faster lookup
    text_set = set(text)
    vocab_set = set(vocabulary)
    
    # Return the set of words in the text but not in the vocabulary
    return text_set - vocab_set

# Example usage:
text = ["apple", "banana", "orange", "apple", "grape"]
vocabulary = ["banana", "grape", "kiwi"]

result = words_not_in_vocab(text, vocabulary)
print("Words in text but not in vocabulary:", result)
```

```python
import nltk
nltk.download('brown')
import re
import nltk
from nltk.corpus import stopwords, brown
from collections import Counter

def find_words_with_frequency(corpus: str, frequency: int=3):

    words = brown.words(categories=corpus)
    word_counts = Counter(words)
    answers = [word for word, count in word_counts.items() if count >= frequency]

    return answers

if __name__ == "__main__":

    #Create stopwords;
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    print(type(stop_words))

    #Create dataset from nltk;
    #Choose a possible type in Brown Corpus, for example 'news';
    corpus_category = 'news'

    #Show answer;
    answers = find_words_with_frequency(corpus_category, 3)
    print(answers)
```

```python
def percent(word, text):
    # Count the total number of words in the text
    total_words = len(text.split())

    # Count the number of occurrences of the given word
    word_count = text.lower().split().count(word.lower())

    # Calculate the percentage
    if total_words > 0:
        percentage = (word_count / total_words) * 100
    else:
        percentage = 0

    return percentage

# Example usage:
text = "This is a sample text with sample words. This text is just for demonstration purposes."
word = "simple"
print(f"The word '{word}' occurs {percent(word, text):.2f}% of the time in the text.")
```

```python
import textacy

def extract_verb_phrases(text):
    # Convert text to a SpaCy doc
    doc = textacy.make_spacy_doc(text, lang='en_core_web_sm')

    # Define a pattern for verb phrases
    pattern = [{"POS": "AUX"}, {"POS": "VERB"}]

    # Find matches using SpaCy's POS-based regex matching
    verb_phrases = textacy.extract.token_matches(doc, pattern)
    
    # Print all Verb Phrases
    for match in verb_phrases:
        print(match.text)

# Example usage:
text = "I may bake a cake for my birthday. The talk will introduce reader about Use of baking"
extract_verb_phrases(text)

```
