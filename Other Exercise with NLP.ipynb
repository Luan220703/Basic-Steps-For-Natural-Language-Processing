```python
import re
import nltk
from nltk.corpus import stopwords, gutenberg

def frequently_words(text: str, stop_words: set, number_words: int=50) -> dict:

    dict_words = {}

    word_list_from_text = re.split(r'[ \n]+', text)

    for word in word_list_from_text:

        if word not in stop_words:

            if word not in dict_words:
                dict_words[word] = 1
            else:
                dict_words[word] += 1

    dict_words = dict(sorted(dict_words.items(), key=lambda item: item[1], reverse=True))
    top_words = {k: v for k, v in list(dict_words.items())[:number_words]}

    return top_words

if __name__ == "__main__":

    #Create stopwords;
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    print(type(stop_words))

    #Create dataset from nltk;
    nltk.download('gutenberg')
    book_names = gutenberg.fileids()
    selected_book = 'austen-emma.txt'
    book_text = gutenberg.raw(selected_book)

    #Show answer;
    answers = frequently_words(book_text, stop_words)
    print(answers)
```

```python
import re
import nltk
from nltk.corpus import stopwords, brown
from collections import Counter

def find_words_with_frequency(corpus: str, frequency: int=3):

    words = brown.words(categories=corpus)
    word_counts = Counter(words)
    answers = [word for word, count in word_counts.items() if count >= frequency]

    return answers

if __name__ == "__main__":

    #Create stopwords;
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    print(type(stop_words))

    #Create dataset from nltk;
    #Choose a possible type in Brown Corpus, for example 'news';
    corpus_category = 'news'

    #Show answer;
    answers = find_words_with_frequency(corpus_category, 3)
    print(answers)
```

```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "My sister has a dog and she loves him"

# Process the text with spaCy
doc = nlp(text)

# Initialize lists to store the replacements
replacements = []
objects = []

# Iterate through the tokens in the processed text
for token in doc:
    # Check if the token is a pronoun
    if token.pos_ == "PRON":
        # Replace pronoun with its antecedent (head noun)
        antecedent = token._.coref_clusters[0].main.text if token._.coref_clusters else token.text
        replacements.append(antecedent)
        objects.append(token.text)
    else:
        replacements.append(token.text)
        objects.append("")

# Print the replacements and objects
print(replacements)
print(objects)
```

```python
def percent(word, text):
    # Count the total number of words in the text
    total_words = len(text.split())

    # Count the number of occurrences of the given word
    word_count = text.lower().split().count(word.lower())

    # Calculate the percentage
    if total_words > 0:
        percentage = (word_count / total_words) * 100
    else:
        percentage = 0

    return percentage

# Example usage:
text = "This is a sample text with sample words. This text is just for demonstration purposes."
word = "sample"
print(f"The word '{word}' occurs {percent(word, text):.2f}% of the time in the text.")
```

```python
import nltk

def tokenize_text(text):
    # Tokenize the text into sentences
    sentences = nltk.sent_tokenize(text)

    # Tokenize each sentence into words
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

    return tokenized_sentences

if __name__ == "__main__":
    # Sample text
    text = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."

    # Tokenize the text
    tokenized_text = tokenize_text(text)

    # Display original string
    print("Original string:")
    print(text)

    # Display sentence-tokenized copy in a list
    print("\nSentence-tokenized copy in a list:")
    print(tokenized_text)

    # Display the sentences in a readable format
    print("\nRead the list:")
    for sentence in tokenized_text:
        print(' '.join(sentence))
```

```python
import re

def custom_tokenize(text):

    sentence_pattern = re.compile(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s*(?=\S)')
    sentences = sentence_pattern.split(text)
    word_pattern = re.compile(r'\b\w+\.?\b')
    tokenized_text = [word_pattern.findall(sentence) + ['.'] for sentence in sentences if sentence]

    return tokenized_text

if __name__ == "__main__":


    #Create dataset;
    text = "Joe waited for the train. The train was late. Mary and Samatha took the bus. I looked for Mary and Samatha at the bus station."

    #Show answer;
    answer = custom_tokenize(text)
    for sentence_tokens in answer:
        print(sentence_tokens)
```
